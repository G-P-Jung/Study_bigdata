# Matplotlib
다앙한 형태의 그래프를 통해서 데이터 시각화를 할 수 있는 라이브러리

```python
import matplotlib.pyplot as plt
```

## 1. 그래프 기본

```python
x = [1,2,3]
y = [2, 4, 8]
plt.plot(x,y)
```

```python
print(plt.plot(x,y))
```

```python
plt.plot(x,y)
plt.show() # 위에 뜨는 글자 없이 깔끔하게 볼 수 있다
```

## Title 설정

```python
plt.plot(x,y)
plt.title("Line graph")
```

```python
plt.plot(x,y)
plt.title("꺾은선 그래프")
```
