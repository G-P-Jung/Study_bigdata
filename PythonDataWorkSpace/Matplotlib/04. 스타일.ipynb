# 4. 스타일

```python
import matplotlib.pyplot as plt
import matplotlib
# matplotlib.rcParams['font.family'] = 'Malgun Gothic' # windows
matplotlib.rcParams['font.family'] = 'AppleGothic' # mac
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시 마이너스 글자가 깨지는 것을 방지
```

```python
x = [1,2,3]
y = [2, 4, 8]
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y, linewidth = 10)
```

```python
plt.plot(x,y, linewidth = 5)
```

## 마커

```python
plt.plot(x,y, marker = 'o')
```

```python
plt.plot(x,y, marker = 'o', linestyle = 'None')
```

```python
plt.plot(x,y, marker = 'v')
```

```python
plt.plot(x,y, marker = 'v', markersize = 10)
```

```python
plt.plot(x,y, marker = 'X', markersize = 10)
```

## google에서 matplotlib marker 검색하면 스타일 알 수 있음

```python
plt.plot(x,y, marker = 'o', markersize = 10, markeredgecolor = 'red')
```

```python
plt.plot(x,y, marker = 'o', markersize = 20, markeredgecolor = 'red')
```

```python
plt.plot(x,y, marker = 'o', markersize = 10, markeredgecolor = 'red', markerfacecolor = 'yellow')
```

## 선 스타일

## matplotlib linestype 검색하면 스타일 찾을 수 있음

```python
plt.plot(x,y, linestyle = ':')
```

```python
plt.plot(x,y, linestyle = '--')
```

```python
plt.plot(x,y, linestyle = '-.')
```

```python
plt.plot(x,y, linestyle = '-')
```

## 색깔

## matplotlib color 검색하면 색상 알 수 있음

```python
plt.plot(x,y, color = 'pink')
```

```python
plt.plot(x,y, color = '#ff0000')
```

```python
plt.plot(x,y, color = 'b')
```

```python
plt.plot(x,y, color = 'g')
```

## 포멧

```python
plt.plot(x,y, 'ro--') # color, marker, linestyle
```

```python
plt.plot(x,y, 'bv:')
```

```python
plt.plot(x,y, 'go') # linestyle = None
```

```python
plt.plot(x,y, color = 'g', marker = 'o', linestyle = 'None')
```

## 축약어

```python
plt.plot(x,y, marker = 'o', mfc = 'red', ms = 10, mec = 'blue', ls = ':')
```

## 투명도

```python
plt.plot(x,y, marker = 'o', mfc = 'red', ms = 10, alpha = 0.3) # alpha = 투명도
```

## 그래프의 크기

```python
plt.figure(figsize = (10,5))
plt.plot(x,y)
```

```python
plt.figure(figsize = (5,10))
plt.plot(x,y)
```

```python
plt.figure(figsize = (10,5), dpi = 50) # dots per inch = 확대
plt.plot(x,y)
```

## 배경색

```python
plt.figure(facecolor = 'yellow')
plt.plot(x,y)
```

```python
plt.figure(facecolor = '#9dbef2')
plt.plot(x,y)
```

```python
plt.figure( figsize = (5,5), facecolor = 'green')
plt.plot(x,y,marker = 'o', linestyle = '--', markeredgecolor = 'red', markersize = 10)
```
